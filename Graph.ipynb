#  Zellophan-Paradoxon & SSNIP-Test – Warum Kartellbehörden aufpassen müssen

## 1. Was macht der SSNIP-Test?
Der SSNIP-Test fragt:
*"Was passiert, wenn wir den Preis um 5–10 % erhöhen? Werden KundInnen auf andere Anbieter oder Märkte ausweichen?"*

- Wenn viele Kundinnen wechseln → *„Es gibt viel Konkurrenz, der Markt ist gross!“*
- Wenn kaum jemand wechselt → *„Es gibt keine guten Alternativen, der Markt ist klein!“*

**Aber Achtung:** Der Test funktioniert nur richtig, wenn der **Ausgangspreis normal ist**.

---

## 2. Das Problem mit Monopolpreisen
Ein Monopolist (z. B. Booking) **optimiert seinen Preis**, um den **höchsten Gewinn** zu machen.

- Wenn der Preis **zu niedrig** ist, verdient er zu wenig.
- Wenn der Preis **zu hoch** ist, verliert er zu viele Kunden.
- **Also setzt er genau den Preis, der für ihn am besten ist.**

➡ Wenn Booking also **bereits ein Monopol hat**, ist der Preis **schon so hoch wie sinnvoll**.

---

## 3. Warum der SSNIP-Test hier falsch liegt
Jetzt kommt der Fehler:
- Der SSNIP-Test sagt: *„Lass uns den Preis 10 % erhöhen und schauen, ob Kunden abwandern.“*
- **Problem:** Der Preis ist **bereits maximal** – eine weitere Erhöhung macht nur Verluste!
- Der Test würde sagen: *„Oh, viele Kunden wechseln, also gibt es viel Konkurrenz!“*

➡ **Das ist falsch!** Die Kunden wechseln nur, weil der Preis vorher schon extrem hoch war.

---

##  4. Zellophan-Paradoxon – Die falsche Marktdefinition
Der SSNIP-Test würde fälschlicherweise sagen:
*"Es gibt viele Alternativen! Booking hat kein Monopol, weil Kunden einfach woanders buchen."*

Aber in Wahrheit:
Kunden **wechseln nicht wegen echter Konkurrenz**, sondern weil die Preise **bereits hoch sind**.
**Der Markt sieht grösser aus, als er wirklich ist.**
**Kartellbehörden könnten dadurch falsche Entscheidungen treffen.**

➡ Dieses Problem nennt man das **Zellophan-Paradoxon**.

---

##  5. Fazit für Kartellrecht & Marktanalyse
- Wenn ein Unternehmen **schon Monopolpreise nimmt**, macht der SSNIP-Test **keinen Sinn mehr**.
- Die Marktdefinition wird **künstlich zu weit gezogen**.
- **Kartellbehörden müssen smartere Methoden verwenden**, um Marktmacht richtig zu erkennen.


---

Monopolisten setzen bereits den optimalen Preis.
Ein SSNIP-Test auf Monopolpreise gibt ein falsches Bild.
Das Zellophan-Paradoxon lässt den Markt **grösser erscheinen, als er ist**.
